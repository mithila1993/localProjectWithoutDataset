```python
import csv
from nltk.tokenize import word_tokenize
import xlrd
import re
from nltk import bigrams
import operator
from collections import Counter
from nltk.corpus import stopwords
import string
```

```python
dataset = []
```

```python
def preProcess():
    global counter, previousText, dataset
    with open('dataset/process/analyze/temp.csv','r',encoding='Latin1')as f:

      data = csv.reader(f)
      for row in data:
        dataset.append(row[1])
```

```python
preProcess()
```

```python
emoticons_str = r"""
    (?:
        [:=;] # Eyes
        [oO\-]? # Nose (optional)
        [D\)\]\(\]/\\OpP] # Mouth
    )"""

regex_str = [
    emoticons_str,
    r'<[^>]+>',  # HTML tags
    r'(?:@[\w_]+)',  # @-mentions
    
]

tokens_re = re.compile(r'(' + '|'.join(regex_str) + ')', re.VERBOSE | re.IGNORECASE)
emoticon_re = re.compile(r'^' + emoticons_str + '$', re.VERBOSE | re.IGNORECASE)


def tokenize(s):
    return tokens_re.findall(s)


def preprocess(s, lowercase=False):
    tokens = tokenize(s)
    if lowercase:
        tokens = [token if emoticon_re.search(token) else token.lower() for token in tokens]
    return tokens

punctuation = list(string.punctuation)
stop = stopwords.words('english') + punctuation + ['The','tourism','RT','I','Tourism','rt', 'via','...']

count_all = Counter()
count_hash = Counter()
count_bigram = Counter()

for x in range(1, 200):
    value = dataset[x]
    #print(value)
    #print(preprocess(value))

    terms_all = [term for term in preprocess(value) if term not in stop]
    print(terms_all)
    
    terms_bigram = bigrams(terms_all)

    #terms_single = set(terms_all)

    terms_hash = [term for term in preprocess(value)
                  if term.startswith('#')]

    count_all.update(terms_all)
    count_hash.update(terms_hash)  
    count_bigram.update(terms_bigram)


    #print("\n")
print('popular terms - ',count_all.most_common(5))
print('popular hash terms - ',count_hash.most_common(5))
print('popular bigrams - ',count_bigram.most_common(5))
```
